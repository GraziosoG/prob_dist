This notebook outputs the graphs of different probability distributions. The probability distributions include discrete uniform (starting from 1), geometric, binomial, negative binomial, hypergeometric, and poisson. The scipy stats library is extensively used. The mean, variance, skewness and kurtosis were also printed for convenience. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Discrete Uniform Distribution (starting from 1)
### Input n: the number of likely outcomes

```python
from scipy.stats import uniform

def make_unif(n):
    X = [i for i in range(1, n+1)]
    unif_probs = [1/n for i in range(1, n+1)]
    print("All", str(n), "data points have the same probability of", str(1/n))
    mean, var = (n+1)/2, (n*n-1)/12
    print("\nmean", str(mean), "\nvariance", str(var))
    fig, ax = plt.subplots(1, 1, figsize=(8, 4))
    ax.plot(X, unif_probs, 'ro', ms=8, label='unif pmf', alpha=0.8)
    
    plt.xlabel("X", fontsize="14")
    plt.ylabel("Probability", fontsize="14")
    plt.xticks(np.arange(1, n+1, 1), fontsize=14)
    plt.yticks(fontsize=14)
    plt.title("Uniform Distribution with n= " + str(n), fontsize="18")
    ax.vlines(X, 0, unif_probs, colors='r', lw=5, alpha=0.5);
```

```python
make_unif(5)
```

# Geometric Distribution
## X is the number of failures before the first success
### Input 1 p: probability of success
### Input 2 x: datapoint range (optional, default = 10)

```python
from scipy.stats import geom

def make_geom(p, x=10):
    X = [i for i in range(1, x+1)]
    geom_probs = geom.pmf(X, p) # Get the probabilities of geometric probability distribution
    for x in range(len(geom_probs)):
        print(x, " ", geom_probs[x])
    mean, var, skew, kurt = geom.stats(p, moments='mvsk')
    print("\nmean", str(mean), "\nvariance", str(var), "\nskewness", str(skew), "\nkurtosis", str(kurt))
    fig, ax = plt.subplots(1, 1, figsize=(10, 6))
    ax.plot([i for i in range(0, x+1)], geom_probs, 'bo', ms=8, label='geom pmf', alpha=0.8)
    
    plt.xlabel("X", fontsize="14")
    plt.ylabel("Probability", fontsize="14")
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.title("Geometric Distribution with p= " + str(p), fontsize="18")
    ax.vlines([i for i in range(0, x+1)], 0, geom_probs, colors='b', lw=5, alpha=0.5);
```

```python
make_geom(0.8)
```

```python
make_geom(0.25, 6)
```

# Binomial Distribution
## X is the number of success in n trials with probability of success p 
### Input 1 n: independent repetitions of trials
### Input 2 p: probability of success

```python
from scipy.stats import binom

def make_bino(n, p):
    X = [i for i in range(0, n+1)]
    binom_probs = binom.pmf(X, n, p)
    for x in range(len(binom_probs)):
        print(x, " ", binom_probs[x])
    mean, var, skew, kurt = binom.stats(n, p, moments='mvsk')
    print("\nmean", str(mean), "\nvariance", str(var), "\nskewness", str(skew), "\nkurtosis", str(kurt))
    fig, ax = plt.subplots(1, 1, figsize=(10, 6))
    ax.plot(X, binom_probs, 'ro', ms=8, label='binom pmf', alpha=0.8)
    
    plt.xlabel("X", fontsize="14")
    plt.ylabel("Probability", fontsize="14")
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    plt.title("Binomial Distribution with n= " + str(n) + " and p= " + str(p), fontsize="18")
    ax.vlines(X, 0, binom_probs, colors='r', lw=5, alpha=0.5);
```

```python
make_bino(6, 0.4)
```

# Negative Binomial Distribution
## X is the number of failures before the rth success, with probability of success p
### Input 1 r: rth success
### Input 2 p: probability of success

```python
from scipy.stats import nbinom

def make_nbin(r, p):
    X = [i for i in range(0, 2*r+3)]
    nbinom_probs = nbinom.pmf(X, r, p)
    for x in range(len(nbinom_probs)):
        print(x, " ", nbinom_probs[x])
    mean, var, skew, kurt = nbinom.stats(r, p, moments='mvsk')
    print("\nmean", str(mean), "\nvariance", str(var), "\nskewness", str(skew), "\nkurtosis", str(kurt))
    fig, ax = plt.subplots(1, 1, figsize=(10, 6))
    ax.plot(X, nbinom_probs, 'bo', ms=8, label='nbinom pmf', alpha=0.8)
    
    plt.xlabel("X", fontsize="14")
    plt.ylabel("Probability", fontsize="14")
    plt.xticks(np.arange(0, 2*r+3, 1),fontsize=14)
    plt.yticks(fontsize=14)
    plt.title("Negative Binomial Distribution with r= " + str(r) + " and p= " + str(p), fontsize="18")
    ax.vlines(X, 0, nbinom_probs, colors='b', lw=5, alpha=0.5);
```

```python
make_nbin(3, 0.5)
```

# Hypergeomtetric Binomial Distribution
### X is the number of type I objects in the chosen group of n
### Input 1 b: number of type I objects
### Input 2 g: number of type II objects
### Input 3 n: chosen group size

```python
from scipy.stats import hypergeom

def make_hgeom(b, g, n):
    if n < b:
        X = [i for i in range(0, n+1)]
    else:
        X = [i for i in range(0, b+1)]
    hgeom_probs = hypergeom.pmf(X, b+g, b, n)
    for x in range(len(hgeom_probs)):
        print(x, " ", hgeom_probs[x])
    mean, var, skew, kurt = hypergeom.stats(b+g, b, n, moments='mvsk')
    print("\nmean", str(mean), "\nvariance", str(var), "\nskewness", str(skew), "\nkurtosis", str(kurt))
    fig, ax = plt.subplots(1, 1, figsize=(10, 6))
    ax.plot(X, hgeom_probs, 'ro', ms=8, label='hgeom pmf', alpha=0.8)
    
    plt.xlabel("X", fontsize="14")
    plt.ylabel("Probability", fontsize="14")
    plt.xticks(np.arange(0, n+1, 1),fontsize=14)
    plt.yticks(fontsize=14)
    plt.title("Hypergeometric Distribution with b= " + str(b) + ", g= " + str(g)+ ", and n= " + str(n), fontsize="18")
    ax.vlines(X, 0, hgeom_probs, colors='r', lw=5, alpha=0.5);
```

```python
make_hgeom(4, 20, 5)
```

```python
make_hgeom(9, 6, 3)
```

# Poisson Distribution
## X is the number of events happened
### Input 1 l: lambda, also the mean
### Input 2 x: datapoint range (optional, default = 5)

```python
from scipy.stats import poisson

def make_pois(l, x=5):
    X = [i for i in range(x+1)]
    pois_probs = poisson.pmf(X, l) # Get the probabilities of geometric probability distribution
    for x in range(len(pois_probs)):
        print(x, " ", pois_probs[x])
    mean, var, skew, kurt = poisson.stats(l, moments='mvsk')
    print("\nmean", str(mean), "\nvariance", str(var), "\nskewness", str(skew), "\nkurtosis", str(kurt))
    fig, ax = plt.subplots(1, 1, figsize=(10, 6))
    ax.plot(X, pois_probs, 'bo', ms=8, label='pois pmf', alpha=0.8)
    
    plt.xlabel("X", fontsize="14")
    plt.ylabel("Probability", fontsize="14")
    plt.xticks(np.arange(0, x+1, 1),fontsize=14)
    plt.yticks(fontsize=14)
    plt.title("Poisson Distribution with lambda= " + str(l), fontsize="18")
    ax.vlines(X, 0, pois_probs, colors='b', lw=5, alpha=0.5);
```

```python
make_pois(3, 8)
```
